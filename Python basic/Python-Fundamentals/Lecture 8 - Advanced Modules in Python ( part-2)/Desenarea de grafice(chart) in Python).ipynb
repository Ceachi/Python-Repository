## Exemplu 1

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3], [8,9,4]) # se definesc pe orizontala punctele din 1,2,3 si corespondente
                            # pe verticala 8,9,4
    
plt.show()
```

## Exemplu 2

```python
import matplotlib.pyplot as plt

x = [1,2,3]
y = [8,9,4]

x2 = [1,2,3]
y2 = [10, 14, 12]

plt.plot(x, y, label = 'Prima linie') 
'''
se numeste dreapta rezultata din
corespondenta coordonatelor x si y
'''

plt.plot(x2, y2, label = 'A doua linie')

plt.xlabel('Axa X') # se eticheteaza axa x avand numele 'Axa X'
plt.ylabel('Axa Y') # similar pentru axa y
plt.title('Afisam Y in functie de X') # se denumeste figura
plt.legend() # se afiseaza legenda
```
